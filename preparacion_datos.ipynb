```python
#importo librerias
import pandas as pd
import json
import ast
```

```python
#creo listas que contendran los datos
lista_reviews = []
lista_items = []
lista_games = []
```

```python
#leo los json
with open("PI MLOps - STEAM/australian_user_reviews.json", 'r', encoding="utf-8")as file:
    for linea in file.readlines():
        lista_reviews.append(ast.literal_eval(linea))

with open("PI MLOps - STEAM/output_steam_games.json", 'r', encoding="utf-8")as file:
    for linea in file.readlines():
        lista_games.append(json.loads(linea))
```

```python
with open("PI MLOps - STEAM/australian_users_items.json", 'r', encoding="utf-8")as file:
    for linea in file.readlines():
        lista_items.append(ast.literal_eval(linea))
```

```python
#creo los dataframe
df_reviews = pd.DataFrame(lista_reviews)
df_items = pd.DataFrame(lista_items)
df_games = pd.DataFrame(lista_games)
```
